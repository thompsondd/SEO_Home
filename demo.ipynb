```python
huyen = ['Nhà Bè', 'Củ Chi','Hóc Môn', 'Bình Chánh']
New = []
for d in df[["Quận (Huyện)","Phường (Xã)","Địa chỉ"]].values:
    q,p,a = d
    if q in huyen:
        q = f"huyện {q}"
        p = f"xã {p}"
    else:
        q = f"quận {q}"
        p = f"phường {p}"
    address=", ".join([a,p,q,"Thành phố Hồ Chí Minh"])
    New.append(address)
ndf=pd.DataFrame(New,columns=["adds"])
ndf
```
